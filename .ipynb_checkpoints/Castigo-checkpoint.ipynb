$\textbf{Non-Probability Sampling (Muestreo no probabilístico)}$

Esta técnica depende de la capacidad del investigador para seleccionar elementos para una muestra, no depende de la probabilidad sino de las condiciones que permiten hacer el muestreo (acceso o disponibilidad, conveniencia, etc); son seleccionadas con mecanismos informales y no aseguran la total representación de la población. 
Hay situaciones en que es más conveniente usar un muestreo no probabilístico, por ejemplo cuando vamos a hacer
estudios de casos, de poblaciones heterogéneas, o en estudios que son dirigidos a poblaciones y grupos muy específicos donde interesa una cuidadosa y controlada selección de sujetos con determinadas características.


Existen 4 ramas dentro de esta tecnica:


$\textbf{Convenience Sampling (Muestreo de conveniencia)}$

El muestreo de conveniencia es un tipo de muestreo en el que la primera fuente de datos primaria disponible se utilizará para la investigación sin requisitos adicionales. En otras palabras, este método de muestreo consiste en buscar a los participantes en donde se puedan encontrar, y por lo general, donde sea conveniente. En el muestreo de conveniencia no se identificaron criterios de inclusión antes de la selección de los sujetos. Todos los sujetos están invitados a participar.
Este método se utiliza cuando la disponibilidad de la muestra es rara y también costosa. Así que en base a la conveniencia se seleccionan muestras.

En su forma básica, el método de muestreo de conveniencia se puede aplicar deteniendo a personas al azar en la calle y haciendo preguntas al cuestionario. 

    Ventajas:
    
    -Simplicidad de muestreo y facilidad de investigación.
    
    -Útil para estudios piloto y generación de hipótesis.
    
    -La recopilación de datos se puede facilitar en un corto período de tiempo.
    
    -Más barato para implementar métodos alternativos de muestreo.
    
    Desventajas:
    
    -Altamente vulnerable a los sesgos e influencias de selección más allá del control del investigador.
    
    -Alto nivel de error de muestreo.
    
    -Los estudios que utilizan muestreo de conveniencia tienen poca credibilidad.
    

$\textbf{Purposive Sampling (Muestreo intencional)}$

También conocido como Juzgado, es una técnica de muestreo en la que el investigador se basa en su propio juicio al elegir a miembros de la población para participar en el estudio, se produce cuando los elementos seleccionados para la muestra son elegidos por el criterio del investigador. Los investigadores a menudo creen que pueden obtener una muestra representativa utilizando un buen juicio, lo que resultará en un ahorro de tiempo y dinero.

En el muestreo intencional, se debe utilizar el juicio personal para elegir casos que ayuden a responder preguntas de investigación o lograr objetivos de investigación.

Según el tipo de casos, el muestreo intencional se puede dividir en las siguientes seis categorías:

Caso típico. Explica casos que son medios y normales.
Caso extremo o desviado. Obtención de muestras de casos que se perciben como inusuales o raros, como explorar las razones del fracaso corporativo al entrevistar a ejecutivos despedidos por accionistas.

 
    -El muestreo de casos críticos se enfoca en casos específicos que son dramáticos o muy importantes.
    
    -El muestreo heterogéneo o de variación máxima se basa en el criterio del investigador para seleccionar participantes con características diversas. Esto se hace para asegurar la presencia de máxima variabilidad dentro de los datos primarios.
    
    -El muestreo homogéneo se centra en un subgrupo particular en el que todos los miembros de la muestra son similares, como una ocupación o nivel en particular en la jerarquía de una organización.
    
    -El muestreo teórico es un caso especial de muestreo intencional que se basa en un método inductivo de teoría fundamentada.
    
Por ejemplo: si queremos comprender el pensamiento de las personas que están interesadas en obtener un título de maestría, entonces el criterio de selección sería "¿Está interesado en obtener una maestría en ...?"

Todas las personas que respondan con un "No" serán excluidas de nuestra muestra.

    Ventajas:
    
    -El muestreo intencional es uno de los métodos de muestreo más rentables y eficientes.
    
    -El muestreo intencional puede ser el único método apropiado disponible si solo hay un número limitado de fuentes de datos primarios que pueden contribuir al estudio
    
    -Esta técnica de muestreo puede ser eficaz para explorar situaciones antropológicas en las que el descubrimiento de un significado puede beneficiarse de un enfoque intuitivo.
    
    Desventajas:
    
    -Vulnerabilidad a los errores de juicio del investigador.
    
    -Bajo nivel de confiabilidad y altos niveles de sesgo.
    
    -Incapacidad para generalizar los resultados de la investigación.

$\textbf{Quota Sampling (Muestreo de cuotas)}$

Este tipo de muestreo depende de algún estándar preestablecido, es un método de muestreo para recopilar datos representativos de un grupo. La aplicación del muestreo por cuotas garantiza que el grupo de muestra represente ciertas características de la población elegida por el investigador.

El muestreo de cuotas se puede dividir en dos grupos: controlado y no controlado.

    -El muestreo de cuotas controlado implica la introducción de ciertas restricciones para limitar la selección de muestras por parte del investigador.

    -El muestreo de cuota no controlado se asemeja al método de Convenience Sampling (muestreo de conveniencia) de una manera que el investigador puede elegir libremente los miembros del grupo de muestra de acuerdo con su voluntad.
    

Ejemplo:

Asumamos que su objetivo de investigación es evaluar el impacto de las diferentes culturas en la motivación de los empleados. Debe evaluar la efectividad de las herramientas de motivación de los empleados teniendo en cuenta las diferencias de género entre la fuerza laboral.

El muestreo de cuotas se puede aplicar de la siguiente manera:

1. División de la población en grupos específicos. Los empleados como marco de muestreo deben dividirse en los siguientes cinco grupos de acuerdo con su fondo cultural:

    -Europeo
    -Asia (India)
    -Asiático (China)
    -Negro (africano)
    -Otro
    
2. Calcular una cuota para cada grupo. Su supervisor confirma que para alcanzar los objetivos de la investigación, serían apropiados 30 representantes de cada grupo y el tamaño total de la muestra de 150 encuestados.

3. Determinar las condiciones específicas que se deben cumplir y la cuota en cada grupo

Ambos sexos, hombres y mujeres deben estar representados por igual en su grupo de muestra. Esta es una condición importante que debe ser satisfecha. En consecuencia, se reclutan 15 hombres y 15 mujeres de cada grupo.

La aplicación del muestreo por cuotas garantiza que el grupo de muestra represente ciertas características de la población elegida por el investigador. En el ejemplo anterior, una representación igual de ambos sexos, hombres y mujeres, ha sido elegida como una característica importante del muestreo.


    Ventajas:
    
    -El muestreo de cuotas surge como una opción atractiva ya que la recolección de datos primarios se puede hacer en un tiempo más corto.

    -La aplicación del muestreo de cuotas puede ser rentable.

    -El muestreo por cuotas no depende de la presencia de los marcos de muestreo. En ocasiones en las que no existe un marco de muestreo adecuado, el muestreo por cuotas puede ser la única opción apropiada disponible.
    
    Desventajas:
    
    -Igual que otros métodos de muestreo no probabilístico, en el muestreo por cuotas no es posible calcular el error de muestreo y la proyección de los resultados de la investigación a la población total es arriesgada.

    -Si bien esta técnica de muestreo puede ser muy representativa de las características que definen la cuota, otras características importantes pueden estar representadas de manera desproporcionada en el grupo final de la muestra.

    -Existe un gran potencial para el sesgo del investigador y la calidad del trabajo puede sufrir debido a la incompetencia del investigador y / o la falta de experiencia.

$\textbf{Snowball Sampling (Muestreo de referencia)}$

Esta técnica se utiliza en situaciones en las que la población es completamente desconocida y rara.

Por lo tanto, tomaremos la ayuda del primer elemento que seleccionamos para la población y le pediremos que recomiende otros elementos que se ajusten a la descripción de la muestra necesaria. Así que esta técnica de referencia continúa, aumentando el tamaño de la población como una bola de nieve.

Por ejemplo, si está estudiando el nivel de satisfacción del cliente de un club privado, le resultará cada vez más difícil encontrar fuentes de datos primarias a menos que un miembro esté dispuesto a proporcionarle contactos de otros miembros.

Este método de muestreo involucra fuentes de datos primarias que nominan a otras fuentes de datos primarias potenciales para ser utilizadas en la investigación. En otras palabras, el método de muestreo de la bola de nieve se basa en las referencias de los sujetos iniciales para generar sujetos adicionales. Por lo tanto, al aplicar este método de muestreo, los miembros del grupo de muestra se reclutan a través de la cadena de referencia.

Existen 3 patrones en esta tecnica:

    1. Muestreo lineal. La formación de un grupo de muestra comienza con un solo sujeto y el sujeto proporciona solo una referencia.
    
    2. Muestreo exponencial no discriminativo. El primer tema reclutado para el grupo de muestra proporciona múltiples referencias.
    
    Muestreo exponencial discriminativo. Los sujetos dan múltiples referencias, sin embargo, solo uno nuevo es reclutado entre ellos.
    
La aplicación del muestreo de bolas de nieve comprende las siguientes etapas:

    -Establecer un contacto con uno o dos casos iniciales desde el marco muestral. Esta etapa suele ser la más difícil.
    
    -Solicitar los casos iniciales para identificar más casos.
    
    -Pedir nuevos casos para identificar otros casos (y así sucesivamente)
    

    Ventajas:
    
    -La capacidad de reclutar poblaciones ocultas.
    
    -La posibilidad de recopilar datos primarios de manera rentable.
    
    -Los estudios se pueden completar en un corto período de tiempo.
    
    -Se requiere muy poca planificación para iniciar el proceso de recolección de datos primarios
    
    Desventajas:
    
    -Sobremuestrear una red particular de pares puede llevar a sesgos.
    
    -Dificultad para iniciar el encadenamineto.
    
    -No hay garantía sobre la representatividad de las muestras.
    
    -No es posible determinar el error de muestreo y hacer inferencias estadísticas de la muestra a la población debido a la ausencia de una selección aleatoria de muestras.
