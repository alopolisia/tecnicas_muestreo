# SAMPLING TECHNIQUES

It is one of the most important factors which determines the accuracy of your research/survey result. If anything goes wrong with your sample then it will be directly reflected in the final result. 


### Population
Population is the collection of the elements which has some or the other characteristic in common.

### Sampling
Sampling is the process of selecting a sample.

### Sample
Sample is the subset of the population.

There are two categories of sampling techniques:
* Probability Sampling
* Non - Probability Sampling

The difference lies between the above two is whether the sample selection is based on randomization or not. With randomization, every element gets equal chance to be picked up and to be part of sample for study.

# Probability Sampling

This category uses __randomization__ to make sure that every element of the population gets an equal chance to be part of the selected sample.

There are 5 techniques in this category:

## Simple Random Sampling
Every element has an __equal chance__ of getting selected to be the part sample. It is used when we don‚Äôt have any kind of prior information about the target population.

For example:
Random selection of 20 students from class of 50 student. Each student has equal chance of getting selected. The probability of selection is 1/50.


## Stratified Sampling
This technique divides the elements of the population into __small subgroups__ (strata) based on the __similarity__ in such a way that the elements within the group are homogeneous and heterogeneous among the other subgroups formed. And then the elements are __randomly selected__ from each of these strata. We need to have prior information about the population to create subgroups.


## Cluster Sampling
The entire population is divided into __clusters or sections__ and then the clusters are randomly selected. All the elements of the cluster are used for sampling. 

Cluster sampling can be done in following ways:

* ### Sing Stage Cluster Sampling
The Entire cluster is selected randomly for sampling.


* ### Two Stage Cluster Sampling
First we randomly select clusters and then from those selected clusters we randomly select elements for sampling.


## Systematic Sampling
In this technique the selection of elements is __systematic__ and not random except the __first element__. All the elements are put together in a sequence first where each element has the equal chance of being selected.

For a sample of size n, we divide our population of size N into subgroups of k elements.

Then we select our __first element randomly__ from the first subgroup of k elements.

To select other elements of sample, perform following:
* The number of elements in each group is k = N/n
* If our first element is n1 then second element = n1 + k and third element = n2 + k and so on...
* Taking an example of N=20, n=5.
* k = 20 / 5, k= 4.
* Now we randomly select first element from the first subgroup.
* n1 = 3, n2 = n1 + k = 3 + 4 = 7, n3 = n2 + k = 7 + 4 = 11.


## Multi Stage Sampling
It is a __combination__ of two or more methods described above.

Population is divided into __multiple clusters__ and then these clusters are __further divided__ and grouped into various sub groups (strata) based on similarity.

One or more clusters can be __randomly selected__ from each stratum. This process continues until the cluster can‚Äôt be divided anymore.

For example country can be divided into states, cities, urban and rural and all the areas with similar characteristics can be merged together to form a strata.


# Non-Probability Sampling

Esta t√©cnica depende de la capacidad del investigador para seleccionar elementos para una muestra, no depende de la probabilidad sino de las condiciones que permiten hacer el muestreo (acceso o disponibilidad, conveniencia, etc); son seleccionadas con mecanismos informales y no aseguran la total representaci√≥n de la poblaci√≥n. 
Hay situaciones en que es m√°s conveniente usar un muestreo no probabil√≠stico, por ejemplo cuando vamos a hacer
estudios de casos, de poblaciones heterog√©neas, o en estudios que son dirigidos a poblaciones y grupos muy espec√≠ficos donde interesa una cuidadosa y controlada selecci√≥n de sujetos con determinadas caracter√≠sticas.


Existen 4 ramas dentro de esta tecnica:


$\textbf{Convenience Sampling (Muestreo de conveniencia)}$

El muestreo de conveniencia es un tipo de muestreo en el que la primera fuente de datos primaria disponible se utilizar√° para la investigaci√≥n sin requisitos adicionales. En otras palabras, este m√©todo de muestreo consiste en buscar a los participantes en donde se puedan encontrar, y por lo general, donde sea conveniente. En el muestreo de conveniencia no se identificaron criterios de inclusi√≥n antes de la selecci√≥n de los sujetos. Todos los sujetos est√°n invitados a participar.
Este m√©todo se utiliza cuando la disponibilidad de la muestra es rara y tambi√©n costosa. As√≠ que en base a la conveniencia se seleccionan muestras.

En su forma b√°sica, el m√©todo de muestreo de conveniencia se puede aplicar deteniendo a personas al azar en la calle y haciendo preguntas al cuestionario. 

    Ventajas:
    
    -Simplicidad de muestreo y facilidad de investigaci√≥n.
    
    -√ötil para estudios piloto y generaci√≥n de hip√≥tesis.
    
    -La recopilaci√≥n de datos se puede facilitar en un corto per√≠odo de tiempo.
    
    -M√°s barato para implementar m√©todos alternativos de muestreo.
    
    Desventajas:
    
    -Altamente vulnerable a los sesgos e influencias de selecci√≥n m√°s all√° del control del investigador.
    
    -Alto nivel de error de muestreo.
    
    -Los estudios que utilizan muestreo de conveniencia tienen poca credibilidad.
    

$\textbf{Purposive Sampling (Muestreo intencional)}$

Tambi√©n conocido como Juzgado, es una t√©cnica de muestreo en la que el investigador se basa en su propio juicio al elegir a miembros de la poblaci√≥n para participar en el estudio, se produce cuando los elementos seleccionados para la muestra son elegidos por el criterio del investigador. Los investigadores a menudo creen que pueden obtener una muestra representativa utilizando un buen juicio, lo que resultar√° en un ahorro de tiempo y dinero.

En el muestreo intencional, se debe utilizar el juicio personal para elegir casos que ayuden a responder preguntas de investigaci√≥n o lograr objetivos de investigaci√≥n.

Seg√∫n el tipo de casos, el muestreo intencional se puede dividir en las siguientes seis categor√≠as:

Caso t√≠pico. Explica casos que son medios y normales.
Caso extremo o desviado. Obtenci√≥n de muestras de casos que se perciben como inusuales o raros, como explorar las razones del fracaso corporativo al entrevistar a ejecutivos despedidos por accionistas.

¬†
    -El muestreo de casos cr√≠ticos se enfoca en casos espec√≠ficos que son dram√°ticos o muy importantes.
    
    -El muestreo heterog√©neo o de variaci√≥n m√°xima se basa en el criterio del investigador para seleccionar participantes con caracter√≠sticas diversas. Esto se hace para asegurar la presencia de m√°xima variabilidad dentro de los datos primarios.
    
    -El muestreo homog√©neo se centra en un subgrupo particular en el que todos los miembros de la muestra son similares, como una ocupaci√≥n o nivel en particular en la jerarqu√≠a de una organizaci√≥n.
    
    -El muestreo te√≥rico es un caso especial de muestreo intencional que se basa en un m√©todo inductivo de teor√≠a fundamentada.
    
Por ejemplo: si queremos comprender el pensamiento de las personas que est√°n interesadas en obtener un t√≠tulo de maestr√≠a, entonces el criterio de selecci√≥n ser√≠a "¬øEst√° interesado en obtener una maestr√≠a en ...?"

Todas las personas que respondan con un "No" ser√°n excluidas de nuestra muestra.

    Ventajas:
    
    -El muestreo intencional es uno de los m√©todos de muestreo m√°s rentables y eficientes.
    
    -El muestreo intencional puede ser el √∫nico m√©todo apropiado disponible si solo hay un n√∫mero limitado de fuentes de datos primarios que pueden contribuir al estudio
    
    -Esta t√©cnica de muestreo puede ser eficaz para explorar situaciones antropol√≥gicas en las que el descubrimiento de un significado puede beneficiarse de un enfoque intuitivo.
    
    Desventajas:
    
    -Vulnerabilidad a los errores de juicio del investigador.
    
    -Bajo nivel de confiabilidad y altos niveles de sesgo.
    
    -Incapacidad para generalizar los resultados de la investigaci√≥n.

$\textbf{Quota Sampling (Muestreo de cuotas)}$

Este tipo de muestreo depende de alg√∫n est√°ndar preestablecido, es un m√©todo de muestreo para recopilar datos representativos de un grupo. La aplicaci√≥n del muestreo por cuotas garantiza que el grupo de muestra represente ciertas caracter√≠sticas de la poblaci√≥n elegida por el investigador.

El muestreo de cuotas se puede dividir en dos grupos: controlado y no controlado.

    -El muestreo de cuotas controlado implica la introducci√≥n de ciertas restricciones para limitar la selecci√≥n de muestras por parte del investigador.

    -El muestreo de cuota no controlado se asemeja al m√©todo de Convenience Sampling (muestreo de conveniencia) de una manera que el investigador puede elegir libremente los miembros del grupo de muestra de acuerdo con su voluntad.
    

Ejemplo:

Asumamos que su objetivo de investigaci√≥n es evaluar el impacto de las diferentes culturas en la motivaci√≥n de los empleados. Debe evaluar la efectividad de las herramientas de motivaci√≥n de los empleados teniendo en cuenta las diferencias de g√©nero entre la fuerza laboral.

El muestreo de cuotas se puede aplicar de la siguiente manera:

1. Divisi√≥n de la poblaci√≥n en grupos espec√≠ficos. Los empleados como marco de muestreo deben dividirse en los siguientes cinco grupos de acuerdo con su fondo cultural:

    -Europeo
    -Asia (India)
    -Asi√°tico (China)
    -Negro (africano)
    -Otro
    
2. Calcular una cuota para cada grupo. Su supervisor confirma que para alcanzar los objetivos de la investigaci√≥n, ser√≠an apropiados 30 representantes de cada grupo y el tama√±o total de la muestra de 150 encuestados.

3. Determinar las condiciones espec√≠ficas que se deben cumplir y la cuota en cada grupo

Ambos sexos, hombres y mujeres deben estar representados por igual en su grupo de muestra. Esta es una condici√≥n importante que debe ser satisfecha. En consecuencia, se reclutan 15 hombres y 15 mujeres de cada grupo.

La aplicaci√≥n del muestreo por cuotas garantiza que el grupo de muestra represente ciertas caracter√≠sticas de la poblaci√≥n elegida por el investigador. En el ejemplo anterior, una representaci√≥n igual de ambos sexos, hombres y mujeres, ha sido elegida como una caracter√≠stica importante del muestreo.


    Ventajas:
    
    -El muestreo de cuotas surge como una opci√≥n atractiva ya que la recolecci√≥n de datos primarios se puede hacer en un tiempo m√°s corto.

    -La aplicaci√≥n del muestreo de cuotas puede ser rentable.

    -El muestreo por cuotas no depende de la presencia de los marcos de muestreo. En ocasiones en las que no existe un marco de muestreo adecuado, el muestreo por cuotas puede ser la √∫nica opci√≥n apropiada disponible.
    
    Desventajas:
    
    -Igual que otros m√©todos de muestreo no probabil√≠stico, en el muestreo por cuotas no es posible calcular el error de muestreo y la proyecci√≥n de los resultados de la investigaci√≥n a la poblaci√≥n total es arriesgada.

    -Si bien esta t√©cnica de muestreo puede ser muy representativa de las caracter√≠sticas que definen la cuota, otras caracter√≠sticas importantes pueden estar representadas de manera desproporcionada en el grupo final de la muestra.

    -Existe un gran potencial para el sesgo del investigador y la calidad del trabajo puede sufrir debido a la incompetencia del investigador y / o la falta de experiencia.

$\textbf{Snowball Sampling (Muestreo de referencia)}$

Esta t√©cnica se utiliza en situaciones en las que la poblaci√≥n es completamente desconocida y rara.

Por lo tanto, tomaremos la ayuda del primer elemento que seleccionamos para la poblaci√≥n y le pediremos que recomiende otros elementos que se ajusten a la descripci√≥n de la muestra necesaria. As√≠ que esta t√©cnica de referencia contin√∫a, aumentando el tama√±o de la poblaci√≥n como una bola de nieve.

Por ejemplo, si est√° estudiando el nivel de satisfacci√≥n del cliente de un club privado, le resultar√° cada vez m√°s dif√≠cil encontrar fuentes de datos primarias a menos que un miembro est√© dispuesto a proporcionarle contactos de otros miembros.

Este m√©todo de muestreo involucra fuentes de datos primarias que nominan a otras fuentes de datos primarias potenciales para ser utilizadas en la investigaci√≥n. En otras palabras, el m√©todo de muestreo de la bola de nieve se basa en las referencias de los sujetos iniciales para generar sujetos adicionales. Por lo tanto, al aplicar este m√©todo de muestreo, los miembros del grupo de muestra se reclutan a trav√©s de la cadena de referencia.

Existen 3 patrones en esta tecnica:

    1. Muestreo lineal. La formaci√≥n de un grupo de muestra comienza con un solo sujeto y el sujeto proporciona solo una referencia.
    
    2. Muestreo exponencial no discriminativo. El primer tema reclutado para el grupo de muestra proporciona m√∫ltiples referencias.
    
    Muestreo exponencial discriminativo. Los sujetos dan m√∫ltiples referencias, sin embargo, solo uno nuevo es reclutado entre ellos.
    
La aplicaci√≥n del muestreo de bolas de nieve comprende las siguientes etapas:

    -Establecer un contacto con uno o dos casos iniciales desde el marco muestral. Esta etapa suele ser la m√°s dif√≠cil.
    
    -Solicitar los casos iniciales para identificar m√°s casos.
    
    -Pedir nuevos casos para identificar otros casos (y as√≠ sucesivamente)
    

    Ventajas:
    
    -La capacidad de reclutar poblaciones ocultas.
    
    -La posibilidad de recopilar datos primarios de manera rentable.
    
    -Los estudios se pueden completar en un corto per√≠odo de tiempo.
    
    -Se requiere muy poca planificaci√≥n para iniciar el proceso de recolecci√≥n de datos primarios
    
    Desventajas:
    
    -Sobremuestrear una red particular de pares puede llevar a sesgos.
    
    -Dificultad para iniciar el encadenamineto.
    
    -No hay garant√≠a sobre la representatividad de las muestras.
    
    -No es posible determinar el error de muestreo y hacer inferencias estad√≠sticas de la muestra a la poblaci√≥n debido a la ausencia de una selecci√≥n aleatoria de muestras.

## Classification Test Problems
Classification is the problem of assigning labels to observations.

### Blobs Classification Problem
The __make_blobs()__ function can be used to generate blobs of points with a Gaussian distribution.

You can control how many blobs to generate and the number of samples to generate, as well as a host of other properties.

The problem is suitable for __linear classification problems__ given the linearly separable nature of the blobs.

The example below generates a 2D dataset of samples with __three blobs__ as a multi-class classification prediction problem. Each observation has __two inputs__ and 0, 1, or 2 class values.

```python
from sklearn.datasets.samples_generator import make_blobs
from matplotlib import pyplot
from pandas import DataFrame

#FUNCI√ìN MAKE BLOBS
#-------------------------------------------------------------------------
# generate 2d classification dataset
X, y = make_blobs(n_samples=100, centers=3, n_features=2)
#-------------------------------------------------------------------------

# scatter plot, dots colored by class value
df = DataFrame(dict(x=X[:,0], y=X[:,1], label=y))

colors = {0:'red', 1:'blue', 2:'green'}
fig, ax = pyplot.subplots()
grouped = df.groupby('label')

for key, group in grouped:
    group.plot(ax=ax, kind='scatter', x='x', y='y', label=key, color=colors[key])
    
pyplot.show()
```

### Moons Classification Problem
The __make_moons()__ function is for binary classification and will generate a swirl(Remolino) pattern, or two moons.

You can control how noisy the moon shapes are and the number of samples to generate.

This test problem is suitable for algorithms that are capable of learning __nonlinear class boundaries__.

The example below generates a moon dataset with moderate noise.

```python
from sklearn.datasets import make_moons
from matplotlib import pyplot
from pandas import DataFrame

#FUNCI√ìN MAKE MOON
#------------------------------------------------------------
# generate 2d classification dataset
X, y = make_moons(n_samples=100, noise=0.1)
#------------------------------------------------------------

# scatter plot, dots colored by class value
df = DataFrame(dict(x=X[:,0], y=X[:,1], label=y))

colors = {0:'red', 1:'blue'}
fig, ax = pyplot.subplots()
grouped = df.groupby('label')

for key, group in grouped:
    group.plot(ax=ax, kind='scatter', x='x', y='y', label=key, color=colors[key])
    
pyplot.show()
```

### Circles Classification Problem

The __make_circles()__ function generates a binary classification problem with datasets that fall into concentric circles.

Again, as with the moons test problem, you can control the amount of noise in the shapes.

This test problem is suitable for algorithms that can learn __complex non-linear manifolds__.

The example below generates a circles dataset with some noise.

```python
from sklearn.datasets import make_circles
from matplotlib import pyplot
from pandas import DataFrame

#MAKE CIRCLE
#------------------------------------------------------------------
# generate 2d classification dataset
X, y = make_circles(n_samples=100, noise=0.05)
#------------------------------------------------------------------

# scatter plot, dots colored by class value
df = DataFrame(dict(x=X[:,0], y=X[:,1], label=y))

colors = {0:'red', 1:'blue'}
fig, ax = pyplot.subplots()
grouped = df.groupby('label')

for key, group in grouped:
    group.plot(ax=ax, kind='scatter', x='x', y='y', label=key, color=colors[key])
    
pyplot.show()
```

## Regression Test Problems
The __make_regression()__ function will create a dataset with a __linear relationship__ between inputs and the outputs.

You can configure the number of samples, number of input features, level of noise, and much more.

This dataset is suitable for algorithms that can learn a __linear regression function__.

The example below will generate 100 examples with one input feature and one output feature with modest noise.

```python
from sklearn.datasets import make_regression
from matplotlib import pyplot

#MKE REGRESION--------------------------------------------------------------
# generate regression dataset
X, y = make_regression(n_samples=100, n_features=1, noise=0.1)
#---------------------------------------------------------------------------

# plot regression dataset
pyplot.scatter(X,y)
pyplot.show()
```

En este caso se aplicara la regresi√≥n, es por eso que necesitaremos encontrar la ecuaci√≥n lineal de dichos datos para obtener una aproximaci√≥n.

La formula de una ecuaci√≥n lineas es la siguiente:

        y = ax + b
        
Es necesario aplicar las siguientes formulas para encontrar los valores de las variables necesarias.

        
$a=\binom{n\sum xy - \sum x\sum y}{n\sum x^{2} - (\sum x)^{2}}$



$b = \binom{\sum y-a\sum x}{n}$




```python
n = 6

x = country_stats['GDP per capita'].tolist()
sumatoriaX = sum(x)

y = country_stats['Life satisfaction'].tolist()
sumatoriaY = sum(y)

xy = [1,2,3,4,5,6]
for i in range(n):
    xy[i] = x[i]*y[i]
    
sumatoriaXY = sum(xy)

for i in range(n):
    x[i] = x[i]*x[i]
sumatoriaX2 = sum(x)

print("Sumatoria de X = ",sumatoriaX)
print("Sumatoria de Y = ",sumatoriaY)
print("Sumatoria de X^2 = ",sumatoriaX2)
print("Sumatoria de XY = ",sumatoriaXY)
```

Una vez que sabemos los valores de las respectivas sumatorias, procederemos a remplazarlas en la ecuaci√≥n.

```python
a = (n*sumatoriaXY - sumatoriaX * sumatoriaY)/(n*sumatoriaX2-(sumatoriaX * sumatoriaX))
print("Valor de A: ",a)
```

```python
b = (sumatoriaY - a * sumatoriaX)/n
print("Valor de B: ",b)
```

```python
print("y = ",a,"x + ",b)
```

```python
def funcion(x):
    return a*x+b
```
